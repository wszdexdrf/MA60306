```python
import pandas as pd

df = pd.read_csv("[ftx] processed_tweets.csv")
df_p = pd.read_csv("prices/ftx_prices_15m.csv")
```

```python
df = df.sort_values('Datetime')
```

```python
from datetime import datetime

i = 0
while i < len(df):
    dt_object = datetime.fromtimestamp(int(df.iloc[i]["Datetime"]))
    if dt_object.minute % 15 == 0:
        break
    i += 1
```

```python
begin = datetime.fromtimestamp(int(df.iloc[i]["Datetime"]))
end = datetime.fromisoformat(df_p.iloc[-1]["dateTime"].split()[0])
```

```python
begin, end
```

```python
x = 15
tlow = int(df.iloc[i]["Datetime"])
thigh = tlow + x * 60
tf = end.timestamp()
tlow_array = []
thigh_array = []
score = []
```

```python
sent, c = 0, 0
while thigh < tf:
    if int(df.iloc[i]["Datetime"]) >= tlow and int(df.iloc[i]["Datetime"]) < thigh:
        sent += df.iloc[i]["polarity"]
        c += 1
        i += 1
    else:
        tlow_array.append(tlow)
        thigh_array.append(thigh)
        score.append(sent)
        tlow = thigh
        thigh = tlow + x * 60
        sent = 0
        c = 0
```

```python
df1 = pd.DataFrame()
df1["time start"] = tlow_array
df1["time high"] = thigh_array
df1["sentiment"] = score
```

```python
df1
```

```python
import numpy as np
senti = np.array(df1["sentiment"])
```

```python
i = 0
begin_val = str(begin).replace(":", "-")
while i < len(df_p) and df_p.iloc[i].dateTime < begin_val:
    print(df_p.iloc[i].dateTime, begin_val)
    i += 1
df_p = df_p[i:]
```
